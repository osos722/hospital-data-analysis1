```python
def pie_chart(values_list,labels_list,this_sick):
  plt.pie(values_list, labels=labels_list, autopct='%1.1f%%', startangle=90)
  plt.title(f'No-show Distribution ({this_sick}  = 0)')
  plt.show()
```
